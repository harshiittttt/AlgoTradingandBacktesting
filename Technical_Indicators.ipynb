#CUMMALTIVE RETURNS,Drawdowns etc

```python
reliance = yf.download("RELIANCE.NS")
```

```python
reliance
```

```python
reliance.columns
```

```python
# chnaging from multindex to flatindex
reliance.columns = reliance.columns.droplevel('Ticker')
reliance
```

```python
reliance.columns
```

```python
#copy the close column from reliance to rel
rel = reliance.loc[:, "Close"]
rel
```

```python
rel
```

```python
rel = rel.to_frame()
rel.head()
```

```python
rel["d-returns"] = np.log(rel.div(rel.shift(1)))
rel
```

```python
rel.dropna(inplace=True)
rel
```

```python
rel["d-returns"].sum()
```

```python
np.exp(rel["d-returns"].sum())
```

```python
rel["cum-returns"] = rel["d-returns"].cumsum().apply(np.exp)
```

```python
rel.head(5000)
```

```python
rel["cum-returns"].plot(figsize=(12,8),title="Growth of 1 rupee in reliance over the years",fontsize=12)
```

```python
# now calculating annual average returns percent of reliance since
rel["d-returns"].mean()*252*100
```

```python
# now risk factor percent
rel["d-returns"].std()*np.sqrt(252)*100
```

```python
#how to calculate drawdowns
rel["cummmax"] = rel['cum-returns'].cummax()
rel
```

```python
rel[["cum-returns","cummmax"]].plot(figsize=(12,8),title="Reliance buy and hold+cummmax",fontsize=12)
```

```python
rel["drawdown"] = rel['cummmax']-rel['cum-returns']
rel
```

```python
rel['drawdown'].max()
```

```python
rel.drawdown.idxmax()
```

```python
rel.loc[(rel.index=='2025-03-04')]
```

```python
#now calculate drowdown percentage
rel['drawdown%'] = (rel["cummmax"]-rel['cum-returns'])/rel['cummmax']*100
```
