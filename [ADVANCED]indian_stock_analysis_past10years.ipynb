```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ticker = ["RELIANCE.NS","TATAMOTORS.NS","TCS.NS","INFY.NS","ADANIENT.NS","M&M.NS"]
```

```python
stocks = yf.download(ticker,start="2015-01-01",end="2025-04-17")
```

```python
stocks
```

```python
# getting the close price
close = stocks.loc[:,"Close"].copy()
close
```

```python
close.dropna()
```

```python
close.info()
```

```python
# now we normalize all the stocks
normalizestock = close.div(close.iloc[0]).mul(100)
```

```python
normalizestock
```

```python
stock = normalizestock

```

```python
stock.plot(figsize=(15,8),fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret = stock.pct_change(periods=1).dropna()
ret
```

```python
summary=ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
summary["mean"] = summary["mean"]*252
summary["std"] = summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(12,8),s=50,fontsize=15)
```

```python
import seaborn as sns
sns.set(style="darkgrid")  # Options: whitegrid, darkgrid, white, dark, ticks
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(12,8),s=50,fontsize=15)
for i in summary.index:
  plt.annotate(i,xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002),size=10)
plt.xlabel("Annual Risk(std)",fontsize=15)
plt.ylabel("Annual Return(Mean)",fontsize=15)
plt.title("Return V/s Risk")
```

## CORRELATION AND COVARIANCE
